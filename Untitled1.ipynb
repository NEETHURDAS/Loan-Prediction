# Chi Square Test

```python
from sklearn.feature_selection import chi2
chi_columns_interest=['Gender','Marital_Status','Number_of_Dependents','Education_Level','Location','Credit_History']
```

```python
#split the dataset
X = data.drop("Approval_Status",axis=1)   #Feature Matrix
y = data["Approval_Status"] 
```

```python
X.head()
```

```python
Z = X[chi_columns_interest]
```

```python
Z.shape
```

```python
y.shape
```

```python
chi_scores = chi2(Z, y)
```

```python
# higher the chi value, higher the importance
plt.figure(figsize=(12,10))
chi_values = pd.Series(chi_scores[0], index=Z.columns)
chi_values.sort_values(ascending=False, inplace=True)
chi_values.plot.bar(fontsize=12)
plt.savefig('chi2.png')
```

```python
pointbiserial,p_value=stats.pointbiserialr(data['Approval_Status'],data['Total_Income'])
# Print the results
print("Point-Biserial Correlation Coefficient :", pointbiserial)
print("p-value :", p_value)
```

```python
pointbiserial_log_totalincome,p_value_log_totalincome=stats.pointbiserialr(data['Approval_Status'],data['log_totalincome'])
# Print the results
print("Point-Biserial Correlation Coefficient :", pointbiserial_log_totalincome)
print("p-value :", p_value_log_totalincome)
```

# SMOTE for treating the imbalanced dataset

```python
from imblearn.combine import SMOTETomek
```

```python
#splitting of the dataset into test and train data
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, test_size=0.3, random_state=15)
```

```python
# Apply SMOTEtomek to the training data
smtom=SMOTETomek(random_state=42)
X_train_smtom, y_train_smtom = smtom.fit_resample(X_train, y_train)
```

# Decision Tree Model

```python
from sklearn.model_selection import GridSearchCV
```

```python
# Define the parameter grid
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [10, 20, 30, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': [None, 'sqrt', 'log2']
}

# Initialize the Decision Tree model
dt = DecisionTreeClassifier(random_state=42)

# Initialize Grid Search with cross-validation
grid_search = GridSearchCV(estimator=dt, param_grid=param_grid,
                           cv=3, n_jobs=-1, verbose=2, scoring='f1_weighted')

# Fit the grid search to the data
grid_search.fit(X_train_smtom, y_train_smtom)

# Get the best parameters and best score
best_params_dt = grid_search.best_params_
best_score_dt = grid_search.best_score_

print("Best Parameters for Decision Tree:", best_params_dt)
print("Best F1 Score for Decision Tree:", best_score_dt)
```

```python
dt = DecisionTreeClassifier(random_state=42,criterion='gini',max_depth=20,min_samples_leaf=2,min_samples_split=10)
dt_model_train=dt.fit(X_train_smtom,y_train_smtom)
dt_imp=dt_model_train.feature_importances_
```

```python
from sklearn.model_selection import cross_validate, StratifiedKFold
stratified_k_fold = StratifiedKFold(n_splits=10)
# Define scoring
scoring = {
    'accuracy': 'accuracy',
    'precision': 'precision',
    'recall': 'recall',
    'f1': 'f1'
}

# Perform cross-validation
scores = cross_validate(dt, X_train_smtom, y_train_smtom, cv=stratified_k_fold, scoring=scoring)

print('Accuracy:', scores['test_accuracy'].mean())
print('Precision:', scores['test_precision'].mean())
print('Recall:', scores['test_recall'].mean())
print('F1 Score:', scores['test_f1'].mean())
```

```python
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, 30, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}

# Initialize the Random Forest model
rf = RandomForestClassifier(random_state=42)

# Initialize Grid Search with cross-validation
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid,
                           cv=3, n_jobs=-1, verbose=2, scoring='f1_weighted')

# Fit the grid search to the data
grid_search.fit(X_train_smtom, y_train_smtom)

# Get the best parameters and best score
best_params_rf = grid_search.best_params_
best_score_rf = grid_search.best_score_

print("Best Parameters (Grid Search):", best_params_rf)
print("Best F1 Score (Grid Search):", best_score_rf)
```

```python
# Confusion Matrix function

def plot_confusion_matrix(cm, classes=None, title='Confusion matrix'):
    """Plots a confusion matrix."""
    if classes is not None:
        sns.heatmap(cm, cmap="YlGnBu", xticklabels=classes, yticklabels=classes, vmin=0., vmax=1., annot=True, annot_kws={'size':50})
    else:
        sns.heatmap(cm, vmin=0., vmax=1.)
    plt.title(title)
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
# Visualizing cm

cm = confusion_matrix(y_test,rf_pred)
cm_norm = cm / cm.sum(axis=1).reshape(-1,1)

plot_confusion_matrix(cm, classes = rf.classes_, title='Confusion matrix')
plt.savefig('confusionmatrix.png')
```
